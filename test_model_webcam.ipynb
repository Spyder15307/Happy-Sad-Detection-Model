```python
import cv2
import time
from tensorflow.keras.models import load_model
```

```python
model = load_model("models\happy_sad_model.h5")
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)  # 0 represents the default camera (you might need to change it if you have multiple cameras)
```

```python
# # while True:
# #     ret, frame = cap.read()
# #     if not ret:
# #         break

# #     # Preprocess the frame (resize, normalize, etc.) based on your model's requirements
# #     # Assuming preprocessed_frame is your input image
# #     preprocessed_frame = cv2.resize(frame, (256, 256))
# #     preprocessed_frame = preprocessed_frame.reshape((1, 256, 256, 3))
    

# #     # Use your model to predict emotions
# #     prediction = model.predict(preprocessed_frame)

# #     # Get the emotion label (happy or sad)
# #     emotion_label = "Happy" if prediction > 0.5 else "Sad"

# #     # Display the result on the frame
# #     cv2.putText(frame, f"Emotion: {emotion_label}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

# #     # Display the frame
# #     cv2.imshow('Webcam Emotion Detection', frame)

# #     # Break the loop if 'q' key is pressed
# #     if cv2.waitKey(1) & 0xFF == ord('q'):
# #         break

# # # Release the webcam and close the OpenCV window
# # cap.release()
# # cv2.destroyAllWindows()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to grayscale for face detection
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        # Draw a rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

        # Extract the face region for your emotion detection
        face_roi = frame[y:y+h, x:x+w]

        # Preprocess the face region (resize and normalize)
        face_roi = cv2.resize(face_roi, (256, 256))
        face_roi = face_roi / 255.0
        face_roi = face_roi.reshape((1, 256, 256, 3))

        # Use your emotion detection model to predict emotions
        prediction = model.predict(face_roi)

        # Get the emotion label (happy or sad)
        emotion_label = "Happy" if prediction > 0.5 else "Sad"

        # Display the emotion label above the rectangle
        cv2.putText(frame, f"Emotion: {emotion_label}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Display the frame
    cv2.imshow('Webcam Emotion Detection', frame)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the OpenCV window
cap.release()
cv2.destroyAllWindows()
# Load the pre-trained face detection model from OpenCV
# face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# # Access the webcam
# cap = cv2.VideoCapture(0)

# # Countdown
# for i in range(3, 0, -1):
#     print(f"Starting in {i} seconds...")
#     time.sleep(1)

# # Capture a single frame
# ret, frame = cap.read()

# # Convert the frame to grayscale for face detection
# gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# # Detect faces in the frame
# faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)

# for (x, y, w, h) in faces:
#     # Draw a rectangle around the detected face
#     cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

#     # Extract the face region for your emotion detection
#     face_roi = frame[y:y+h, x:x+w]

#     # Preprocess the face region (resize and normalize)
#     face_roi = cv2.resize(face_roi, (256, 256))
#     face_roi = face_roi / 255.0
#     face_roi = face_roi.reshape((1, 256, 256, 3))

#     # Use your emotion detection model to predict emotions
#     prediction = model.predict(face_roi)

#     # Get the emotion label (happy or sad)
#     emotion_label = "Happy" if prediction > 0.5 else "Sad"

#     # Display the emotion label above the rectangle
#     cv2.putText(frame, f"Emotion: {emotion_label}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# # Display the frame with the detected face
# cv2.imshow('Webcam Emotion Detection', frame)
# cv2.waitKey(0)

# # Release the webcam and close the OpenCV window
# cap.release()
# cv2.destroyAllWindows()
```
